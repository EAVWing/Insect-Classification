# ETL Process

This notebook will show the method of transforming the downloaded jpg files into a more usable npz format. While doing so I will also transform the image so they are all the same size.  Most of the images are already 384 x 256 so that is the size I will use.

tqdm is a neat little package that gives a progress bar so you can see how fast the script is going through the file and how many it has done so far.



```python
import glob
import os
import numpy as np
import tqdm
import matplotlib.pyplot as plt
import skimage.transform
```

```python
IMG_SIZE = (384, 256)

input_data = list(glob.glob(pathname='C:/Users/sewing/OneDrive - Do My Own/bug_getter/**/*.jpg'))

for image_path in tqdm.tqdm(input_data):
    image_pixels = plt.imread(image_path)
    resized_image_pixels = skimage.transform.resize(image_pixels, IMG_SIZE)
    image_basepath, _ = os.path.splitext(image_path)
    np.savez(image_basepath+'.npz', pixels=resized_image_pixels, compressed=True)
```
